# Distribution of Gender across company

```python
gender_counts=data.groupby('Department')['Gender'].value_counts().reset_index(name='Count')
```

```python
pivot_data = gender_counts.pivot_table(index='Department', columns='Gender', values='Count', aggfunc='sum', fill_value=0)
```

```python
pivot_data
```

```python

pivot_data.plot(kind='bar',stacked=True,figsize=(10,6))
plt.legend(title='Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
```

# Percetange of each ethnicity by Department 

```python
Ethnicity_by_department=data.groupby('Department')['Ethnicity'].value_counts().reset_index(name='Count')
```

```python
Ethnicity_by_department_pivot=Ethnicity_by_department.pivot_table(index='Department',columns='Ethnicity',values='Count',aggfunc='sum')
```

```python
percentage_data = Ethnicity_by_department_pivot.div(Ethnicity_by_department_pivot.sum(axis=1), axis=0) * 100
```

```python
#attrition rate by DistanceFromHome
attrition_rate_by_distance = data.groupby('DistanceFromHome (KM)')['Attrition'].mean()
plt.figure(figsize=(8, 5))
plt.scatter(attrition_rate_by_distance.index, attrition_rate_by_distance, color='b')
plt.title("Attrition Rate by Distance From Home")
plt.xlabel("Distance From Home (KM)")
plt.ylabel("Attrition Rate")
plt.grid(True)
plt.show()

# Line Chart
plt.figure(figsize=(8, 5))
plt.plot(attrition_rate_by_distance.index, attrition_rate_by_distance, marker='o', color='r')
plt.title("Attrition Rate by Distance From Home")
plt.xlabel("Distance From Home (KM)")
plt.ylabel("Attrition Rate")
plt.grid(True)
plt.show()
```

```python
Attrition_by_BusinessTravel=data.groupby('BusinessTravel')['Attrition'].sum().sort_values(ascending=True).reset_index()
Attrition_by_BusinessTravel
```

```python
sns.barplot(x='BusinessTravel',y='Attrition',data=Attrition_by_BusinessTravel,color='g')
plt.title('Attrition_by_BusinessTravel')
```

# Predictive Analysis

```python
data['OverTime']=data['OverTime'].map({'Yes':1,'No':0})
correlation=data[['Salary','OverTime','YearsAtCompany']].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title("Correlation Matrix of Key Factors and Attrition")
plt.show()
```
